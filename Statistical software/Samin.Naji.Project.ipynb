[Matplotlib](https://matplotlib.org/) was created by neurobiologist John Hunter to work with EEG data. It grew to be used and developed by many people in many different fields. John's goal was that Matplotlib make easy things easy and hard things possible.

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. It can produce publication-quality figures in various formats and platforms, and can be used in scripts, shells, web applications, and graphical user interface toolkits

## Below we describe several common approaches to plotting with Matplotlib:
 
### The pyplot API (application programming interface): 
matplotlib.pyplot is a collection of functions that make Matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

### The object-oriented API:
At its core, Matplotlib is object-oriented. We recommend directly working with the objects, if you need more control and customization of your plots.

In many cases you will create a Figure and one or more Axes using pyplot.subplots and from then on only work on these objects. However, it's also possible to create Figures explicitly (e.g. when including them in GUI applications).

### The pylab API (discouraged)
*Warning:*
Since heavily importing into the global namespace may result in unexpected behavior, the use of pylab is strongly discouraged. Use matplotlib.pyplot instead.

pylab is a module that includes matplotlib.pyplot, numpy, numpy.fft, numpy.linalg, numpy.random, and some additional functions, all within a single namespace. Its original purpose was to mimic a MATLAB-like way of working by importing all functions into the global namespace. This is considered bad style nowadays.

[Matplotlib.pyplot.bar](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#examples-using-matplotlib-pyplot-bar) is a function in the matplotlib library that can be used to create bar plots in Python. A bar plot is a type of graph that shows the values of a variable or a category using rectangular bars of different heights or lengths.

```python
years = (2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020)
salaries = (63000, 65000, 68000, 70000, 69000, 70000, 72000, 76000, 80000, 84000)
prices = (40000, 45000, 44000, 52000, 55000, 53000, 60000, 64000, 62000, 59000)


plt.bar(years, salaries, color ='r')
plt.bar(years, prices, color = 'g')

# plt.xticks(years, ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
# plt.xticks([])


plt.show()
```

[Matplotlib.pyplot.barh](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html) is a function in the matplotlib library that can be used to create horizontal bar plots in Python. A horizontal bar plot is a type of graph that shows the values of a variable or a category using rectangular bars of different lengths or widths. The function takes similar parameters as matplotlib.pyplot.bar, but the x and y axes are swapped. 

```python
label = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
label.reverse()

plt.barh(years, salaries, height = 0.4, color = 'c', edgecolor = 'm', tick_label = label)

plt.grid()
plt.show()
```

[Matplotlib.pyplot.pie](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html) is a function in the matplotlib library that can be used to create pie charts in Python. A pie chart is a type of graph that shows the proportions of a whole using circular sectors of different angles and sizes. 
The wedges are plotted counterclockwise, by default starting from the x-axis.

```python
market_share = [52, 25, 12, 8]
companies = ['apple', 'samsung', 'huawei', 'others']
colors= ['g', 'c', 'm', 'y']

plt.pie(market_share, labels= companies, colors= colors, shadow= True, radius= 0.8, explode= [0, 0.2, 0, 0], autopct= '%.1f%%', startangle= 90 ,counterclock= False)
plt.legend(labels= companies, ncol= 2)
plt.show()
```

```python
days = ('Mon', 'Tue', 'Wen', 'Thu', 'Fri')

sleeping = (7, 8, 6, 11, 7)
eating = (2, 3, 4, 3, 2)
working = (7, 8, 7, 2, 2)
reading =  (8, 5, 7, 8, 13)

plt.stackplot(days, sleeping, eating, working, reading)

plt.xlabel('Days')
plt.ylabel('Hours')
plt.legend(['sleeping', 'eating', 'working', 'reading'])

plt.show()
```

# project 2: (t-tests with examples in Python)
 A t test is a way to compare the means of two groups and see if they are significantly different. There are different types of t tests, depending on the situation and the data.

### One-sample t-test:
 This test compares the mean of a sample to a known or hypothesized value. For example, you can use this test to check if the average height of students in a class is equal to 170 cm. To perform a one-sample t-test in Python, you can use the scipy.stats.ttest_1samp() function. The syntax is: <br>
**from scipy import stats** <br>
**stats.ttest_1samp(sample, popmean)** <br>
where sample is an array-like object containing the sample data, and popmean is the known or hypothesized population mean. The function returns a tuple of the t-statistic and the p-value.


```python
from scipy import stats

sample = [12, 14, 16, 17, 13, 15, 18, 19, 14, 16]

t_stat, p_value = stats.ttest_1samp(sample, 15)

print(f'T-statistic: {t_stat:.2f}')
print(f'P-value: {p_value:.4f}')

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis; there is a significant difference between the sample mean and the hypothesized population mean.")
else:
    print("Fail to reject the null hypothesis; there is no significant difference between the sample mean and the hypothesized population mean.")
```

### Two-sample t-test: 
This test compares the means of two independent samples from different populations. For example, you can use this test to check if the average weight of males and females in a population is different. To perform a two-sample t-test in Python, you can use the scipy.stats.ttest_ind() function. The syntax is: <br>
**from scipy import stats** <br>
**stats.ttest_ind(sample1, sample2, equal_var=True)** <br>
where sample1 and sample2 are array-like objects containing the sample data from the two populations, and equal_var is a boolean argument that indicates whether the two populations have equal variances or not. If equal_var is set to False, the function performs a Welch’s t-test, which does not assume equal variances. The function returns a tuple of the t-statistic and the p-value.


```python
from scipy import stats

group1 = [172, 175, 178, 180, 182, 185, 187, 189, 190, 192]
group2 = [162, 164, 166, 168, 170, 172, 174, 176, 178, 180]

t_stat, p_value = stats.ttest_ind(group1, group2)

print(f'T-statistic: {t_stat:.2f}')
print(f'P-value: {p_value:.4f}')

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis; there is a significant difference between the sample mean and the hypothesized population mean.")
else:
    print("Fail to reject the null hypothesis; there is no significant difference between the sample mean and the hypothesized population mean.")
```

### Paired t-test: 
This test compares the means of two related samples from the same population. For example, you can use this test to check if the average blood pressure of patients before and after a treatment is different. To perform a paired t-test in Python, you can use the scipy.stats.ttest_rel() function. The syntax is: <br>
**from scipy import stats** <br>
**stats.ttest_rel(sample1, sample2)** <br>
where sample1 and sample2 are array-like objects containing the paired sample data from the same population. The function returns a tuple of the t-statistic and the p-value.

```python
from scipy import stats

before = [120, 125, 130, 135, 140, 145, 150, 155, 160, 165]
after = [115, 120, 125, 130, 135, 140, 145, 150, 155, 160]

t_stat, p_value = stats.ttest_rel(before, after)

print(f'T-statistic: {t_stat:.2f}')
print(f'P-value: {p_value:.4f}')

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis; there is a significant difference between the sample mean and the hypothesized population mean.")
else:
    print("Fail to reject the null hypothesis; there is no significant difference between the sample mean and the hypothesized population mean.")
```
